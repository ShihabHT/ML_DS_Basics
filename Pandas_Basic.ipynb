# Panda Series
A Series is essentially a single column of data. It consists of two main components: an array of values and an associated array of labels, called the index. Think of it as a single spreadsheet column with a row number for each entry.

  - 1-Dimensional: A Series can only hold one column of data.
  - Homogeneous: All the elements in a Series must be of the same data type (e.g., all integers, all strings, or all floats).
  - Index: Each value in a Series has a unique label, called the index, which can be an integer, a string, or a date.

```python
# series1 = pd.Series([1, 2, 3, 4, 5])
# series1
series2 = pd.Series(['cat', 'dogesh', 'bird', 'deer'], index=["R1", "R2", "R3", "R4"])
# series2 = pd.Series(['cat', 'dogesh', 'bird', 'deer'])
# series2.index = ["R1", "R2", "R3", "R4"]        # another way
series2
```

# Panda DataFrame
A DataFrame is a two-dimensional data structure, much like a spreadsheet or a SQL table. It's the most common and powerful data structure in Pandas. It's essentially a collection of Series objects that share the same index.

- 2-Dimensional: A DataFrame has both rows and columns.
- Heterogeneous: Each column (or Series) in a DataFrame can have a different data type. This allows you to store a variety of data, like names (strings), ages (integers), and salaries (floats), all within the same DataFrame.

- Axis: DataFrames have two axes: a row index (axis 0) and a column index (axis 1).

In short, a DataFrame is a collection of Series. You can imagine a DataFrame as a dictionary where the keys are the column names and the values are the corresponding Series objects.

```python
n1 = np.array([[100, 200, 300], [400, 500, 600], [700, 800, 900]])
df1 = pd.DataFrame(n1, columns=['C1', 'C2', 'C3'], index=["R1", "R2", "R3"])
df1
```

```python
# Creating data frame using dictionary
d = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'Salary': [50000, 60000, 70000]}
df2 = pd.DataFrame(d, index=['R1', 'R2', 'R3'])
df2
```
